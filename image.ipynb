```python
import requests

OPENROUTER_API_KEY = ""

headers = {
    "Authorization": f"Bearer {OPENROUTER_API_KEY}",
    "Content-Type": "application/json"
}

image_prompt = """
Professional LinkedIn banner about AI in Healthcare.
Modern corporate design, doctors using AI,
data analytics dashboard, clean layout,
high quality, suitable for LinkedIn.
"""

payload = {
    "model": "black-forest-labs/flux-1-dev",
    "prompt": image_prompt,
    "size": "1024x1024"
}

response = requests.post(
    "https://openrouter.ai/api/v1/images/generations",
    headers=headers,
    json=payload
)

print(response.json())
```
